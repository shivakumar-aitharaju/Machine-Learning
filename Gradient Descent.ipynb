# Example 1 (Simple cost function)

## $f(x) = x^2 + x + 1$

```python
def f(x):
    return x ** 2 + x + 1
```

```python
## Creating data

x_1 = np.linspace(start = -3 , stop = 3 , num = 500)
#x_1
```

```python
# Let's plot
plt.xlim(-3 , 3)
plt.ylim(0, 8)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('f(x)', fontsize = 16 )
plt.plot(x_1 , f(x_1))
```

## Slope  and Derivatives

```python
def df(x):
    return 2 * x + 1
```

```python
#  plot function and derivative side by side
plt.figure(figsize=(15,5))

# When you create a subplot, you should specify 3 numbers : row , column and index

plt.subplot(1 , 2 , 1) 
# so.. 1 row , 2cols and 1st index

# 1st Chart cost function
plt.xlim(-3 , 3)
plt.ylim(0, 8)

plt.title('Cost Function' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('f(x)', fontsize = 16 )

plt.plot(x_1 , f(x_1) , color = 'blue' , linewidth = 5)


# 2nd Chart Derivative
plt.subplot(1 , 2, 2)

plt.xlim(-2 , 3)
plt.ylim(-3, 6)

plt.title('Slope of the cost function' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('df(x)', fontsize = 16 )

# Apply a grid
plt.grid()

plt.plot(x_1 , df(x_1) , color = 'skyblue' , linewidth = 5)
# so.. 1 row , 2cols and 2nd index

plt.show()
```

```python
# Here the Gradient Descent

new_x = 3
prev_x = 0

step_multiplier = 0.1

precision = 0.00001


x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    prev_x = new_x 
    grad = df(prev_x)
    new_x = prev_x - step_multiplier * grad

    step_size = abs(new_x - prev_x)
    #print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))
    if step_size < precision:
        print("Loop run times : " , n)
        break

print('Local minimum : ' , new_x)
print("slope of df(x) value at this point is : " , df(new_x))
print("f(x) value or cost at this point is :" , f(new_x))
```

```python
#Superimpose the gradient descent calculations on plot function
plt.figure(figsize=(20,5))

# When you create a subplot, you should specify 3 numbers : row , column and index

plt.subplot(1 , 3 , 1) 
# so.. 1 row , 2cols and 1st index

# 1st Chart cost function
plt.xlim(-3 , 3)
plt.ylim(0, 8)

plt.title('Cost Function' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('f(x)', fontsize = 16 )

plt.plot(x_1 , f(x_1) , color = 'blue' , linewidth = 5 , alpha = 0.7)

values = np.array(x_list)
# s = size of data point
# aplha = transparency
plt.scatter(x_list , f(values) , color = 'red' , s = 100 , alpha = 0.6)

# 2nd Chart Derivative
plt.subplot(1 , 3, 2)
# so.. 1 row , 2cols and 2nd index


#change values in xlim and ylim to zoom in and zoom out our chart
plt.xlim(-2 , 3)
plt.ylim(-3, 6)

plt.title('Slope of the cost function' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('df(x)', fontsize = 16 )

# Apply a grid
plt.grid()

plt.plot(x_1 , df(x_1) , color = 'skyblue' , linewidth = 5 , alpha = 0.6)
plt.scatter(x_list , slope_list , color = 'red' , s = 100 , alpha = 0.5)



# 3rd Chart Derivative (Zoomed in )
plt.subplot(1 , 3, 3)
# so.. 1 row , 2cols and 2nd index

plt.xlim(-0.55 , -0.2)
plt.ylim(-0.3, 0.8)

plt.title('Gradient Descent close up' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)

plt.grid()

plt.plot(x_1 , df(x_1) , color = 'skyblue' , linewidth = 6 , alpha = 0.8)
plt.scatter(x_list , slope_list , color = 'red' , s = 300 , alpha = 0.6)

plt.show()
```

## Example 2 : Multiple minima vs Initial guess & Advanced functions

## $$ g(x) = x^4 - 4x^2 + 5 $$

```python
# Make some data 
x_2 = np.linspace(-2 , 2 , 1000)

# g(x) function and dg(x) function

def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
#  plot function and derivative side by side
plt.figure(figsize=(15,5))

# When you create a subplot, you should specify 3 numbers : row , column and index

plt.subplot(1 , 2 , 1) 
# so.. 1 row , 2cols and 1st index

# 1st Chart cost function
plt.xlim(-2 , 2)
plt.ylim(0.5, 5.5)

plt.title('Cost Function' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('g(x)', fontsize = 16 )

plt.plot(x_2 , g(x_2) , color = 'blue' , linewidth = 5)


# 2nd Chart Derivative
plt.subplot(1 , 2, 2)

plt.xlim(-2 , 2)
plt.ylim(-6, 8)

plt.title('Slope of the cost function' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('dg(x)', fontsize = 16 )

# Apply a grid
plt.grid()

plt.plot(x_2 , dg(x_2) , color = 'skyblue' , linewidth = 5)
# so.. 1 row , 2cols and 2nd index

plt.show()
```

## Gradient Descent as a PYTHON Function

```python
def gradient_descent(derivative_func , initial_guess , multiplier = 0.02 , precision = 0.001 ):
    # ctrl + ] (Select the code and press it for changing the indentation)

    new_x = initial_guess

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(500):
        prev_x = new_x 
        grad = derivative_func(prev_x)
        new_x = prev_x - multiplier * grad

        step_size = abs(new_x - prev_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        if step_size < precision:
            break
    
    return new_x , x_list , slope_list 
```

```python
local_min , list_x , deriva_list = gradient_descent(dg , 0.5 , 0.02 , 0.001)
print('Local min occurs at' , local_min )
print("Num of steps", len(list_x))
```

```python
local_min , list_x , deriva_list = gradient_descent(derivative_func= dg , initial_guess= -0.5 , multiplier= 0.01 , precision= 0.0001)
print('Local min occurs at : ' , local_min )
print("Num of steps : ", len(list_x))
```

```python
local_min , list_x , deriva_list = gradient_descent(derivative_func= dg , initial_guess= -0.1)
print('Local min occurs at : ' , local_min )
print("Num of steps : ", len(list_x))
```

```python
# Calling gradient Descent function
local_min , list_x , deriva_list = gradient_descent(derivative_func= dg , initial_guess= 0.1)
#  plot function and derivative and scatter plot side by side
plt.figure(figsize=(15,5))

# When you create a subplot, you should specify 3 numbers : row , column and index

plt.subplot(1 , 2 , 1) 
# so.. 1 row , 2cols and 1st index

# 1st Chart cost function
plt.xlim(-2 , 2)
plt.ylim(0.5, 5.5)

plt.title('Cost Function' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('g(x)', fontsize = 16 )

plt.plot(x_2 , g(x_2) , color = 'blue' , linewidth = 5 , alpha = 0.8)
plt.scatter(list_x , g(np.array(list_x)) , color = 'red' , s = 100, alpha = 0.6)

# 2nd Chart Derivative
plt.subplot(1 , 2, 2)

plt.xlim(-2 , 2)
plt.ylim(-6, 8)

plt.title('Slope of the cost function' , fontsize = 17)
plt.xlabel('X' , fontsize = 16)
plt.ylabel('dg(x)', fontsize = 16 )

# Apply a grid
plt.grid()

plt.plot(x_2 , dg(x_2) , color = 'skyblue' , linewidth = 5 , alpha = 0.6)
plt.scatter(list_x , deriva_list , color = 'red' , s = 100 , alpha = 0.5)
# so.. 1 row , 2cols and 2nd index

plt.show()
```
