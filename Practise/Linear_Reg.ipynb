```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('dark_background')
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
plt.figure(figsize=(8,5))
plt.title('Year wise house prices')
plt.xlabel('Year')
plt.ylabel('Per capita income (US$)')
plt.scatter(df['year'] , df['per capita income (US$)'], color = 'red', marker = '+' )
```

```python
from sklearn  import linear_model

reg = linear_model.LinearRegression()

# Fitting the data: means you're training the Linear Regression model using the available data points

reg.fit(df[['year']] , df['per capita income (US$)'])
```

```python
# If you jus enter (any year) it give error so try 2d array like [[2020]]

reg.predict([[2030]])
```

## y = mx + b
### m is the coefficiant and b is the intercept  




```python

reg.coef_ # m (slope or (Gradient))
```

```python
reg.intercept_ # b (Y intercept)
```

## How it actually predicts :
### $y = mx + b$
###  x is the value that we want to predict for eg : 2025
### Here y = 828.46507522 * 2025 + -1632210.7578554575


```python
828.46507522 * 2030 + -1632210.7578554575
```

```python
years = pd.read_csv('years.csv')
years.head()
```

```python
p = reg.predict(years)
```

```python
years['per capita income'] = p
years.head()

# If you want to export the csv file
# years.to_csv("prediction.csv" , index = False)
```
